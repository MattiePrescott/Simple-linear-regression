```python
# This will import libraries needed
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# This imports the data file
task_data = pd.read_csv("insurance.csv")
task_data.head()
```

```python
# This assigns the x & y values from rows in the data file
x = task_data.iloc[:,0].values
y = task_data.iloc[:,6].values

# This creates the graph using data from the selected rows
plt.scatter(x,y,color = "b")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

```python
# This will fit a model to the data and will make predictions based off it
charges_model = LinearRegression()

age_1 = [[35]]
age_2 = [[40]]
age_3 = [[45]]

X = task_data["age"].values.reshape(-1,1)
y = task_data["charges"].values

charges_model.fit(X,y)
y_prediction = charges_model.predict(X)

print("Insurance cost for a 35 year old should be:", charges_model.predict(age_1))
print("Insurance cost for a 40 year old should be:", charges_model.predict(age_2))
print("Insurance cost for a 45 year old should be:", charges_model.predict(age_3))
```

```python
# This creates a scatter graph showing line of best fit
x_prediction = x.reshape(-1,1)
y_prediction = charges_model.predict(x_prediction)

plt.scatter(x,y,color = "b")
plt.plot(x_prediction,y_prediction,color = "r")
plt.xlabel("Age of customer")
plt.ylabel("Insurance charge")
plt.show()
```
